## Case study on Presentation of data

##### Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Importing dataset into python environment

```python
df=pd.read_csv('cars_data.csv')
df.head(10)
```

##### Summary of features in the data

```python
df.info()
```

##### Count of null values present in the data

```python
df.isnull().sum()
```

##### Bar plot of male vs female buyers participated in the sales.

```python
freq=df['Buyer Gender'].value_counts()
freq
```

```python
plt.figure(figsize=(8,8))
plt.bar(freq.index,freq.values,width=0.4)
plt.xlabel('Buyer gender')
plt.ylabel('Number of buyers')
plt.title('Male vs Female buyers participated in the sales')
plt.xticks()
plt.yticks()
```

##### Top 5 cars based on sales price
